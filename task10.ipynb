```python
# task 11 Reverse a number using while loop.
n=int(input("enter the numbe:"))
rev=0
while n>0:
    digit=n%10
    rev=rev*10+digit
    n=n//10
print(rev)    
```

```python
# task 12 Check if a number is prime using while loop.
n=int(input("enter the number:"))
i=2
count=0
while i<n//2:
    if n%i==0:
        count+=1
    i+=1
if n>1 and count==0:
    print("prime")
else:
    print("not prime")
```

```python
n = int(input("Enter a number: "))

i = 2
count = 0

while i <= n // 2:
    if n % i == 0:
        count += 1
    i += 1

if n > 1 and count == 0:
    print("Prime Number")
else:
    print("Not Prime Number")
```

```python
# Print Fibonacci series up to n terms using while loop.
#Explanation: Use two variables a=0, b=1 and generate next terms.
n=int(input("enter the number:"))
a=0
b=1
count=0
while count<n:
    print(a)
    c=a+b
    a=b
    b=c
    count+=1
    
    
    
```
