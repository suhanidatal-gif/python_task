```python
#Easy Level (1–10)
#1.
#Write a lambda function to return the square of a number.
#Example:
#Input: 5
#Output: 25
squar=lambda x:x**2 
print(squar(2))
```

```python
# 2 Write a lambda function to return the cube of a number.
#Example:
#Input: 4
#Output: 64

cube=lambda x:x**3
print(cube(2))
```

```python
# 3 Write a lambda function to add two numbers.
#Example:
#Input: 10, 20
#Output: 30
add=lambda x,y:x+y
print(add(10,20))
```

```python
# 5 Write a lambda function that returns "Even" if a number is even; otherwise "Odd".
#Input: 9
#Output: Odd
ret=lambda x:"even" if x%2==0 else "odd"
print(ret(3))
```

```python
# 6 Use map() with lambda to square every element of a list.
#Input:
#[1,2,3,4]

#Output:
#[1,4,9,16]
num=[1,2,3,4]
square=list(map(lambda x:x**2,num))
print(square)

```

```python
#7 Use filter() with lambda to keep only even numbers.
#Input:
#[1,2,3,4,5,6]
#Output:
#[2,4,6]
number=[1,2,3,4,5,6]
result=list(filter(lambda x:x%2==0,number))
print(result)
```

```python
# 8 Use filter() with lambda to keep numbers greater than 50.
#Input:
#[20,45,55,70]
#Output:
#[55,70]

num=[20,45,55,70]
res=list(filter(lambda x :  x>=50,num ))
print(res)

                
```

```python
# 9.Sort a list of numbers in ascending order using lambda.
# [9,5,1,8,2]
sor=[9,5,1,8,2]
result=sorted(sor,key =lambda x:x)
print(result)
```

```python
# 10 Sort a list of numbers in descending order using lambda.
# [9,5,1,8,2]
result=sorted(sor,key =lambda x:x,reverse=True)
print(result)
```

```python
# 11 Use reduce() with lambda to find the sum of all numbers.
#[10,20,30,40]
from functools import reduce 
summ=[10,20,30,40]
result1=reduce(lambda x,y:x+y,summ)
print(result1)
```

```python
# 29 Sort a list of dictionaries by age ascending and then salary descending.
employees = [
{"name":"Ram","age":25,"salary":50000},
{"name":"Amit","age":25,"salary":60000},
{"name":"Neha","age":23,"salary":45000},
{"name":"Raj","age":27,"salary":55000}
]
resul=sorted(employees,key=lambda x:(x["age"],-x["salary"]))
print(resul)
```
