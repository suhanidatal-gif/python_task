```python
# 6 class product

class Product:
    def __init__(self, price, stock):
        self.__price = price
        self.__stock = stock

    def update_price(self, new_price):
        if new_price >= 0:
            self.__price = new_price
            print("Price updated")
        else:
            print("Price cannot be negative")

    def update_stock(self, quantity):
        if self.__stock + quantity >= 0:
            self.__stock += quantity
            print("Stock updated")
        else:
            print("Stock cannot become negative")

    def get_details(self):
        print("Price:", self.__price)
        print("Stock:", self.__stock)


product = Product(1000, 10)

product.get_details()

product.update_price(1200)
product.update_stock(5)

product.get_details()

product.update_stock(-20)
```

```python
# 7 Class Name: Car
class Car:
    def __init__(self, model, year):
        self.__model = model
        self.__year = year

    def get_model(self):
        return self.__model

    def set_model(self, model):
        self.__model = model

    def get_year(self):
        return self.__year

    def set_year(self, year):
        if year <= 2026:
            self.__year = year
            print("Year updated")
        else:
            print("Year cannot be greater than current year")


car = Car("Honda", 2020)

print("Model:", car.get_model())
print("Year:", car.get_year())

car.set_model("Toyota")
car.set_year(2025)

print("Updated Model:", car.get_model())
print("Updated Year:", car.get_year())

car.set_year(2030)
```

```python
# 9 Class Name: Rectangle
class Rectangle:
    def __init__(self, length, width):
        self.__length = length
        self.__width = width

    def set_length(self, length):
        if length > 0:
            self.__length = length
            print("Length updated")
        else:
            print("Length must be positive")

    def get_length(self):
        return self.__length

    def set_width(self, width):
        if width > 0:
            self.__width = width
            print("Width updated")
        else:
            print("Width must be positive")

    def get_width(self):
        return self.__width

    def area(self):
        return self.__length * self.__width


rectangle = Rectangle(10, 5)

print("Length:", rectangle.get_length())
print("Width:", rectangle.get_width())
print("Area:", rectangle.area())

rectangle.set_length(-10)
rectangle.set_width(8)

print("New Area:", rectangle.area())
```

```python
# 10 Class Name: Account
class Account:
    history = []

    def __init__(self, balance):
        self.__balance = balance
        Account.history.append(self.__balance)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            Account.history.append(self.__balance)
            print("Deposit successful")
        else:
            print("Invalid amount")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            Account.history.append(self.__balance)
            print("Withdrawal successful")
        else:
            print("Invalid withdrawal")

    def get_balance(self):
        return self.__balance


account = Account(1000)

account.deposit(500)
account.withdraw(200)

print("Current Balance:", account.get_balance())
print("Balance History:", Account.history)
```

```python
class Temperature:
    def __init__(self, celsius):
        self.__celsius = celsius

    def set_celsius(self, celsius):
        self.__celsius = celsius

    def get_celsius(self):
        return self.__celsius

    def to_fahrenheit(self):
        return (self.__celsius * 9 / 5) + 32


temperature = Temperature(25)

print("Celsius:", temperature.get_celsius())
print("Fahrenheit:", temperature.to_fahrenheit())

temperature.set_celsius(30)

print("Updated Celsius:", temperature.get_celsius())
print("Fahrenheit:", temperature.to_fahrenheit())
```
