```python
# hard 2
n = int(input("Enter number: "))

num = 2

while num <= n:
    i = 2
    prime = True

    while i < num:
        if num % i == 0:
            prime = False
            break
        i += 1

    if prime:
        print(num)

    num += 1
```

```python
# amstrong number
start = int(input("Enter start: "))
end = int(input("Enter end: "))

while start <= end:
    n = start
    temp = n
    sum = 0

    while temp > 0:
        digit = temp % 10
        sum = sum + digit ** 3
        temp = temp // 10

    if sum == n:
        print(n)

    start += 1
```

```python
# task number guessing number using while loop

import random

num = random.randint(1, 100)

chance = 1

while chance <= 7:
    guess = int(input("Enter guess: "))

    if guess == num:
        print("Congratulations!")
        break
    elif guess > num:
        print("Too High")
    else:
        print("Too Low")

    chance += 1

if chance > 7:
    print("Number was", num)
```
