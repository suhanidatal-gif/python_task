```python
# 16 Create a function calculate(a, b, operation) and show how keyword args improve
#readability.
def calcu(a,b,opera):
    if opera =="add":
        print("result:",a+b)
    elif opera =="sub":
        print("result:",a-b)
    elif opera=="div":
        print("result:",a*b)
    else:
        print("invalid opera")
calcu(a=10,b=5,opera="add") 
calcu(a=10,b=5,opera="sub")
```
