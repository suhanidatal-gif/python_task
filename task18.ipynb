```python
# 30 
class Rectangle:
    def __init__(self, length, width):
        self.__length = length
        self.__width = width

    # Getters
    def get_length(self):
        print(self.__length)

    def get_width(self):
        print(self.__width)

    # Setters
    def set_length(self, length):
        self.__length = length

    def set_width(self, width):
        self.__width = width

    # Area method
    def area(self):
        print(self.__length * self.__width)


rectangle = Rectangle(10, 5)

print("Length:")
rectangle.get_length()

print("Width:")
rectangle.get_width()

print("Area:")
rectangle.area()

rectangle.set_length(20)
rectangle.set_width(10)

print("New Area:")
rectangle.area()
```

```python
# 2.Create class BankAccount with private __balance. Implement setter for
#balance with these rules:
#• Balance cannot go below minimum balance (1000).
#• If balance > 100000, give 5% bonus automatically.
#• Log every update (use a class list to store history). Add getter and class
#method to show total accounts created.

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    # Getter
    def get_balance(self):
        print(self.__balance)

    # Setter
    def set_balance(self, balance):

        if balance < 1000:
            print("Balance cannot be below 1000")

        else:
            if balance > 100000:
                bonus = balance * 5 / 100
                balance = balance + bonus
                print("5% bonus added")

            else:
                balance = balance

            self.__balance = balance
            print("Balance updated successfully")


account = BankAccount(50000)

print("Old Balance:")
account.get_balance()

account.set_balance(120000)

print("New Balance:")
account.get_balance()

account.set_balance(500)
```

```python
# 3Create class Student with private __marks (list of 5 subjects). Add setter
# for marks with conditions:





class Student:
    def __init__(self, marks):
        self.__marks = marks

    # Getter
    def get_marks(self):
        print(self.__marks)

    # Setter
    def set_marks(self, marks):

        if len(marks) != 5:
            print("Enter marks for exactly 5 subjects")

        else:
            if any(mark < 0 or mark > 100 for mark in marks):
                print("Marks must be between 0 and 100")

            else:
                old_average = sum(self.__marks) / 5
                new_average = sum(marks) / 5

                if new_average < old_average:
                    print("Average cannot decrease")

                else:
                    self.__marks = marks

                    if any(mark < 40 for mark in marks):
                        print("Marks updated successfully")
                        print("Status: Failed")

                    else:
                        print("Marks updated successfully")
                        print("Status: Passed")


student = Student([70, 80, 90, 75, 85])

print("Old Marks:")
student.get_marks()

student.set_marks([80, 85, 95, 90, 88])

print("New Marks:")
student.get_marks()

student.set_marks([30, 20, 40, 50, 60])
```

```python
# 4Create class Product with private __price and __discount. In setter for
# price:


class Product:
    def __init__(self, price):
        self.__original_price = price
        self.__price = price

    # Getter
    def get_price(self):
        print(self.__price)

    # Setter
    def set_price(self, price):

        if price < self.__original_price * 80 / 100:
            print("Price cannot be lower than 80% of original price")

        else:
            if price > 5000:
                discount = price * 15 / 100
                self.__price = price - discount
                print("15% discount applied")

            else:
                self.__price = price
                print("Price updated successfully")


product = Product(10000)

print("Original Price:")
product.get_price()

product.set_price(9000)

print("New Price:")
product.get_price()

product.set_price(7000)

print("Current Price:")
product.get_price()
```

```python
# 5 Create class User with private __password and __age. In setter methods:
#• Password must be at least 8 characters and contain at least 1 digit.
#• Age cannot be decreased (once set).
#• If age >= 60, automatically set senior_citizen = True (class attribute tracking).
#Add update method that updates both with validation.


class User:
    def __init__(self, password, age):
        self.__password = password
        self.__age = age

    # Getter
    def get_age(self):
        print(self.__age)

    # Update method
    def update_user(self, password, age):

        if len(password) < 8:
            print("Password must contain at least 8 characters")

        else:
            if not any(char.isdigit() for char in password):
                print("Password must contain at least 1 digit")

            else:
                if age < self.__age:
                    print("Age cannot be decreased")

                else:
                    self.__password = password
                    self.__age = age

                    print("User updated successfully")

                    if age >= 60:
                        print("Senior Citizen")

                    else:
                        print("Not a Senior Citizen")


user = User("suhani123", 21)

print("Old Age:")
user.get_age()

user.update_user("newpass123", 22)

print("New Age:")
user.get_age()

user.update_user("abc", 25)

user.update_user("newpassword123", 20)
```
